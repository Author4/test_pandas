```python
import pandas as pd

orders = pd.read_csv('cohort_analytics/orders.csv', index_col='order_id')
orders['order_created_dt'] = pd.to_datetime(orders.order_created_dt)
orders['payment_mounth'] = orders['order_created_dt'].dt.to_period('M')
orders = (
    orders
    .groupby('user_id')
    .agg(first_payment_at = ('order_created_dt', 'min'))
    .merge(orders, how='inner', left_index=True, right_on='user_id')
)
orders['first_payment_mounth'] = orders['first_payment_at'].dt.to_period('M')

# orders = (
#     orders
#     .groupby(['first_payment_mounth', 'payment_mounth'])
#     .agg(revenue = ('order_amount', 'sum'))
# )

#Строим когортный анализ через таблицу pivot_table
orders = (
    orders.pivot_table(
    index='first_payment_mounth',
    columns='payment_mounth',
    values='order_amount',
    aggfunc='sum')
)
#orders.sum(axis='rows')
orders



```
